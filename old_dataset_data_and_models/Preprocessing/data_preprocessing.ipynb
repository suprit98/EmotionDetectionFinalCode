```python
# function to create dataframe from txt files having name like training_data_joy.txt .. in 'Data found on internet' folder

def get_data_to_pandas_dataframe(filenames):
    lines = []
    ratings = []
    emotions = []
    df = pd.DataFrame(columns=["tweets","emotion","ratings"])
    for filename in filenames:
        file = open(filename, encoding="utf8")
        for line in file:
            line = line.lstrip(digits)
            line = line.lstrip()
            rating = line[-6:]
            rating = rating[:-1]
            line_tmp = line[:-6]
            line_tmp = line_tmp.rstrip()
            #print(line_tmp)
            #print("-------------------")
            emotion = ""
            j=0
            for i in range(len(line_tmp)-1,0,-1):
                if line_tmp[i].isspace() == True:
                    j=i
                    break
                emotion+=line_tmp[i]
            #print("hello")
            
            emotion = emotion [::-1]
            #print(emotion)
            line = line[:j]
            line = line.rstrip()
            lines.append(line)
            ratings.append(rating)
            emotions.append(emotion)
        df_tmp = pd.DataFrame(list(zip(lines,emotions,ratings)),columns=["tweets","emotion","ratings"])
        #print(df_tmp)
        df = df.append(df_tmp,ignore_index=True)
    return df
```
