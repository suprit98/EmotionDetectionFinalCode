```python
tokenized_tweet = df['content'].apply(lambda x: x.split()) # tokenizing

model_w2v = gensim.models.Word2Vec(
            tokenized_tweet,
            size=200, # desired no. of features/independent variables 
            window=5, # context window size
            min_count=2,
            sg = 1, # 1 for skip-gram model
            hs = 0,
            negative = 10, # for negative sampling
            workers= 2, # no.of cores
            seed = 34)

model_w2v.train(tokenized_tweet, total_examples= len(df['sentiment']), epochs=20)
```

```python
model_w2v.wv.most_similar(positive="food")
```

```python
#tokenized_tweet=tokenized_tweet.values()
print(tokenized_tweet.values)
```

```python
def word_vector(tokens, size):
    vec = np.zeros(size).reshape((1, size))
    count = 0.
    for word in tokens:
        try:
            vec += model_w2v[word].reshape((1, size))
            count += 1.
        except KeyError: # handling the case where the token is not in vocabulary
                         
            continue
    if count != 0:
        vec /= count
    return vec
```

```python
wordvec_arrays = np.zeros((len(tokenized_tweet), 200))

for i in range(len(tokenized_tweet)):
    wordvec_arrays[i,:] = word_vector(tokenized_tweet.values[i], 200)
    
wordvec_df = pd.DataFrame(wordvec_arrays)
print(wordvec_df.shape)
```

```python
# splitting data into training and validation set
wordvec_train, wordvec_test, ytrain, ytest = train_test_split(wordvec_df, df['sentiment'],  
                                                          random_state=42, 
                                                          test_size=0.2)
```

```python
lreg = LogisticRegression()
lreg.fit(wordvec_train, ytrain) 

prediction_test = lreg.predict(wordvec_test)


print(f1_score(ytest, prediction_test,average=None))
print(accuracy_score(ytest, prediction_test))
```
